### Divide Blog Files into Post Files

```python
def extract_post(input_path):
    '''Extract posts from blog files and return a list of posts from the same author
    Argument:
    input_path: string of absolute path to a single blog file that contains posts from an author
    Return: 
    A list of posts from the author
    '''
    done = False # this marker is used to loop through a blog file and end the loop
    tag_tuples = ('<Blog>','<date>','<post>','</post>','</Blog>')
    post_list = []
    post = ''
    
    with open(input_path, errors="ignore") as in_file:
        while not done: #loop over lines in inuput file and write to output file
            try:
                line = next(in_file).strip()
            except StopIteration: # if reaches the end of the file
                done = True
                break
            except UnicodeDecodeError: # if the current line cannot be decoded
                continue
                
            if line.startswith(tag_tuples): # if the current line is a tag instead of actual post
                if len(post) > 0:
                    post_list.append(post)
                    post = ''
            else:
                post = post + ' ' + line
    return post_list
```

```python
def remove_blank(post_list):
    for post in post_list:
        if re.match('^\s+$', post):
            post_list.remove(post)
    return post_list
```

```python
def extract_label(file_name):
    label_list = file_name.split('.')
    return label_list
```

```python
def combine_text_label(post_list, label_list):
    return [[post] + label_list for post in post_list]
```

```python
# Create a list of file path from raw blogs path
blog_path_list = []
for root, directory, files in os.walk(raw_data_path):
    for f in files:
        if '.xml' in f:
            blog_path_list.append(os.path.join(root, f))
```

```python
# Create post list that will later be put into a dataframe
# This step will take a couple minutes, get some coffee and come back later.
col_names = ['post','id','gender','age','industry','sign']
list_df_post = []

for blog in blog_path_list:
    print(os.path.splitext(os.path.basename(blog))[0])
    label_list = extract_label(os.path.splitext(os.path.basename(blog))[0])
    post_list = extract_post(blog) # extract post into a list of posts
    post_list = remove_blank(post_list) # remove all the items that doesn't contain a post(blank posts)
    df_post_author = pd.DataFrame(combine_text_label(post_list, label_list), columns=col_names)
    list_df_post.append(df_post_author)
```

```python
df_posts = pd.concat(list_df_post).reset_index()
df_posts.drop('index', axis=1, inplace=True)
```

```python
df_posts.to_pickle(intermediate_path +'/post_df.pickle')
```
